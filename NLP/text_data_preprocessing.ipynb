Text Data Cleaning

```python
import pandas as pd
import string
```

```python
df = pd.read_csv('/content/drive/MyDrive/Learning /NLP/Twitter Sentiments.csv')
```

```python
df.head()
```

```python
# ddroping the columns
df = df.drop(columns=['id','label'],axis=1)
df.head()
```

```python
# convert to lower case
df['clean_text'] = df['tweet'].str.lower()
df.sample(5)
```

```python
string.punctuation
```

```python
# removeal of punctuations
def remove_punctuation(text):
  punctuations = string.punctuation
  return text.translate(str.maketrans('','',punctuations))
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: remove_punctuation(x))
df.head()
```

```python

import nltk
from nltk.corpus import stopwords
# downloading stopwords
nltk.download('stopwords')
",".join(stopwords.words('english'))
```

```python
from pickle import STOP
# remove stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
  return " ".join(word for word in str(text).split() if word  not in STOPWORDS)
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: remove_punctuation(x))
df.head()
```

```python
## removel of frequent words
from collections import Counter
word_count = Counter()
for text in df['clean_text']:
  for word in text.split():
    word_count[word]+=1

word_count.most_common(10)
```

```python
Frequent_words = set(word for (word,wc) in word_count.most_common(3))
def remove_freq_words(text):
  return " ".join(word for word in str(text).split() if word not in Frequent_words)
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: remove_freq_words(x))
df.head()
```

```python
# remove Rare words
Rare_words = set(word for (word,wc) in word_count.most_common()[:-10:-1])
Rare_words
```

```python
def remove_rare_words(text):
  return " ".join(word for word in str(text).split() if word not in Rare_words)
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: remove_rare_words(x))
df.head()
```

```python
# removal of special characters
import re
def remove_spl_chars(text):
  text = re.sub('[^a-zA-Z0-9]',' ',text)
  text  = re.sub('\s+',' ',text)
  return text
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: remove_spl_chars(x))
df.head()
```

## Stemming

```python
from nltk.stem.porter import PorterStemmer
st = PorterStemmer()
def stem_words(text):
  return " ".join(st.stem(word) for word in text.split())
```

```python
df['stem_text'] = df['clean_text'].apply(lambda x: stem_words(x))
df.head()
```

## Lematization & POS Tagging


```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')

lemmatizer = WordNetLemmatizer()
nltk.download('averaged_perceptron_tagger')
```

```python
# creating a dictionary
wordnet_map = {"N" : wordnet.NOUN,"V" : wordnet.VERB,"J" : wordnet.ADJ,"R" : wordnet.ADV}
def lemmatize_words(text):
  # find pos tags
  pos_text = pos_tag(text.split())
  return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN))for word, pos in pos_text])

```

```python
df['lematized_text'] = df['clean_text'].apply(lambda x: lemmatize_words(x))
df.sample(5)
```
